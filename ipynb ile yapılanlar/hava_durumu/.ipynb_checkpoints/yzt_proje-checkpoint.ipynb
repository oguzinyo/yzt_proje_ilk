```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("munich.csv")
```

```python
df
```

```python
#bunların hepsi tek bir sütun olduğu için bölmemiz lazım
```

```python
df[['time', 'precipitation_sum (mm)', 'snowfall_sum (cm)']] = \
    df['time;precipitation_sum (mm);snowfall_sum (cm)'].str.split(';', expand=True)
```

```python
df
```

```python
#eski sütunu silelim
```

```python
df = df.drop(columns=['time;precipitation_sum (mm);snowfall_sum (cm)'])
```

```python
df
```

```python
#tip dönüşümlerini yapalım tarihi tarih olarak kaydettik
#sayılar da string di integer a çevirdik
```

```python
df['time'] = pd.to_datetime(df['time'])
df['precipitation_sum (mm)'] = pd.to_numeric(df['precipitation_sum (mm)'], errors='coerce')
df['snowfall_sum (cm)'] = pd.to_numeric(df['snowfall_sum (cm)'], errors='coerce')
```

```python
df
```

```python
df.describe()
```

```python
#en fazla yağış alan 10 güne bakalım
```

```python
#bu günler aykırı hava olayları olarak değerlendirilebilir
```

```python
fazlayagıs_df=df.nlargest(10,"precipitation_sum (mm)")
```

```python
fazlayagıs_df
```

```python
fazlayagıs_df.plot(x="time",y="precipitation_sum (mm)",kind="bar")
```

```python
#en fazla kar yağışı alan 10 güne bakalım
```

```python
#bu günler aykırı hava olayları olarak değerlendirilebilir
```

```python
fazlakar_df= df.nlargest(10,"snowfall_sum (cm)")
```

```python
fazlakar_df
```

```python
fazlakar_df.plot(x="time",y="snowfall_sum (cm)",kind="bar")
```

```python
#iklim trenlerini inceleyelim
```

```python
#bunun için önce index i time olarak ayarlayalım
```

```python
df.set_index('time', inplace=True)
```

```python
df=df.dropna()
```

```python
df
```

```python
aylik_toplam_yagis = df['precipitation_sum (mm)'].resample('ME').sum()
aylik_toplam_kar = df['snowfall_sum (cm)'].resample('ME').sum()
#burada günlük verileri aylık olarak grupladık
#ve ME kullanarak ay sonu month end olarak ayarladık
```

```python
aylik_trend_df = pd.DataFrame({
    'precipitation_sum (mm)': aylik_toplam_yagis,
    'snowfall_sum (cm)': aylik_toplam_kar
})
#burada da yeni bi dataframe oluşturduk
```

```python
aylik_toplam_yagis
```

```python
aylik_toplam_kar
```

```python
aylik_trend_df
```

```python
aylik_toplam_yagis.plot(x="precipitation_sum (mm)",y="time")
```

```python
aylik_toplam_kar.plot(x="snowfall_sum (cm)",y="time")
```
